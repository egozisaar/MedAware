## Read data

```python
with open('data/task1_input','r') as file:
    unfiltered_data = file.readlines()

unfiltered_data
```

## Filter noise

```python
import re
    

date_regex = re.compile('^\s*(3[01]|[12][0-9]|0?[1-9])\-(1[012]|0?[1-9])\-((?:19|20)\d{2})\s*$')
hour_regex = re.compile('(?:[01]\d|2[0123]):(?:[012345]\d):(?:[012345]\d)')

filtered_data = []
for row in unfiltered_data:
    parts = row.split(" ")
    if len(parts) < 3:
        continue
        
    date_candidate = parts[0]
    hour_candidate = parts[1]
    content = ' '.join(parts[2:])[:-1]
    if (date_regex.match(date_candidate) is not None) and (hour_regex.match(hour_candidate) is not None):
        filtered_data.append([date_candidate,hour_candidate,content])

filtered_data
```

## Write to output file

```python
import csv

with open("data/targets.csv","w+") as file:
    csv_writer = csv.writer(file,delimiter=',')
    csv_writer.writerows(filtered_data)
```
